# Feature Engineering & Data Preprocessing

## Objective

The objective of this notebook is to prepare the cleaned customer churn dataset for machine learning.

The following preprocessing steps are performed:

- Create a Machine Learning dataset
- Remove unnecessary and leakage columns
- Separate features and target variable
- Encode categorical variables
- Split data into training and testing sets
- Scale numerical features (where required)

The final output of this notebook will be a machine learning-ready dataset that can be used to build churn prediction models.

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("../data/cleaned/Telco_customer_churn_clean.csv")
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

# Creating Machine Learning Dataset

Not all columns are suitable for predictive modeling.

Columns such as CustomerID and Churn Reason either uniquely identify customers or reveal information that would only be known after churn occurs.

These columns are removed to avoid data leakage and improve model generalization.

```python
columns_to_drop = [
    "CustomerID",
    "City",
    "Zip Code",
    "Latitude",
    "Longitude",
    "Lat Long",
    "Churn Score",
    "CLTV",
    "Churn Reason",
    "Churn Label"
]

ml_df = df.drop(columns=columns_to_drop)
```

```python
ml_df.columns
```

```python
ml_df.shape
```

```python
ml_df.to_csv(
    "../data/processed/ml_dataset.csv",
    index=False
)
```

# Separating Features and Target Variable

The target variable for this project is:

**Churn Value**

0 → Customer Retained

1 → Customer Churned

```python
X = ml_df.drop("Churn Value", axis=1)

y = ml_df["Churn Value"]
```

```python
X.head()
```

```python
y.head()
```

# Encoding Categorical Variables

Machine learning algorithms require numerical input.

Categorical features are converted into numerical values using Label Encoding.

```python
encoder = LabelEncoder()
```

```python
categorical_columns = X.select_dtypes(include="object").columns

categorical_columns
```

```python
for col in categorical_columns:
    X[col] = encoder.fit_transform(X[col])
```

```python
X.head()
```

# Train-Test Split

The dataset is divided into training and testing subsets.

Training Set → 80%

Testing Set → 20%

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)
```

```python
print(X_train.shape)
print(X_test.shape)
```

# Feature Scaling

Numerical features are standardized using StandardScaler.

Scaling improves the performance of algorithms such as Logistic Regression.

```python
scaler = StandardScaler()
```

```python
numerical_columns = [
    "Tenure Months",
    "Monthly Charges",
    "Total Charges"
]
```

```python
X_train[numerical_columns] = scaler.fit_transform(
    X_train[numerical_columns]
)

X_test[numerical_columns] = scaler.transform(
    X_test[numerical_columns]
)
```

```python
X_train.head()
```

```python
X_train.to_csv("../data/processed/X_train.csv", index=False)

X_test.to_csv("../data/processed/X_test.csv", index=False)

y_train.to_csv("../data/processed/y_train.csv", index=False)

y_test.to_csv("../data/processed/y_test.csv", index=False)
```
