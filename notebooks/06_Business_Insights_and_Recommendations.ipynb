# Executive Summary

Customer churn is one of the biggest challenges faced by subscription-based businesses.

In this project, customer demographics, subscription details, service usage, and billing information were analysed to identify the factors influencing churn.

Three machine learning models were developed and evaluated for churn prediction.

Among them, Logistic Regression achieved the best overall performance and was selected as the final predictive model.

The analysis identified several customer segments that are more likely to churn, allowing businesses to implement proactive retention strategies before customers leave.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/cleaned/Telco_customer_churn_clean.csv")

feature_importance = pd.read_csv(
    "../data/processed/feature_importance.csv"
)
```

# Major Business Findings

```python
churn = df["Churn Label"].value_counts()

plt.figure(figsize=(6,5))

plt.bar(
    churn.index,
    churn.values
)

plt.title("Overall Customer Churn")

plt.xlabel("Customer Status")

plt.ylabel("Number of Customers")

plt.show()
```

### Business Insight

Approximately **26% of customers have churned**, indicating a significant loss of recurring revenue.

Although most customers remain active, reducing churn even by a few percentage points could substantially improve long-term profitability.

```python
plt.figure(figsize=(10,6))

plt.barh(
    feature_importance["Feature"][:10],
    feature_importance["Importance"][:10]
)

plt.gca().invert_yaxis()

plt.title("Top Features Influencing Customer Churn")

plt.xlabel("Feature Importance")

plt.show()
```

### Business Insight

Machine learning identified the following variables as the strongest predictors of customer churn:

- Monthly Charges
- Tenure Months
- Contract Type
- Total Charges
- Payment Method

These variables should be prioritised while designing customer retention initiatives.

# High-Risk Customer Profile

Customers most likely to churn typically have the following characteristics:

- Month-to-Month Contract
- Short Tenure
- High Monthly Charges
- Fiber Optic Internet Service
- Electronic Check Payment Method
- Limited Value-Added Services

```python
revenue_risk = df[df["Churn Label"]=="Yes"]["Monthly Charges"].sum()

print(f"Estimated Monthly Revenue at Risk : ${revenue_risk:,.2f}")
```

### Business Insight

Customer churn directly impacts recurring monthly revenue.

Retaining high-risk customers can significantly improve long-term profitability while reducing customer acquisition costs.

# Predictive Model Findings

Three machine learning models were developed and compared.

| Model | Accuracy | Precision | Recall | ROC-AUC |
|--------|---------:|----------:|--------:|---------:|
| Logistic Regression | 80.34% | 65.02% | 56.15% | 84.65% |
| Decision Tree | 73.39% | 49.87% | 51.87% | 66.58% |
| Random Forest | 79.91% | 65.85% | 50.53% | 83.57% |

Logistic Regression was selected as the final predictive model due to its superior balance between Accuracy, Recall, F1 Score, and ROC-AUC.

The model can be integrated into customer relationship management systems to identify customers likely to churn before they leave.

# Strategic Recommendations

## 1. Promote Long-Term Contracts

Encourage Month-to-Month customers to migrate to annual contracts by offering loyalty discounts and bundled service plans.

---

## 2. Strengthen Early Customer Engagement

Customers with shorter tenure are more likely to churn.

## 3. Review Pricing Strategy

Offer personalised pricing plans and retention discounts for high-value customers.

---

## 4. Bundle Value-Added Services

Promote Online Security, Tech Support, Device Protection, and Backup services to increase customer engagement.

---

## 5. Deploy Predictive Churn Monitoring

Integrate the Logistic Regression model into CRM systems to identify high-risk customers and initiate retention campaigns before churn occurs.
