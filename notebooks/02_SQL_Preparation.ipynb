# Customer Churn Prediction & Retention Analytics Platform

## SQL Database Preparation

### Objective

Load the cleaned customer churn dataset into a SQLite database.

The database will be used for SQL-based business analysis in the next phase of the project.

### Output

- SQLite Database (`churn.db`)
- Table (`customer_churn_data`)

```python

import pandas as pd
import sqlite3
```

```python
sql_df = pd.read_csv(
    "../data/cleaned/Telco_customer_churn_clean.csv"
)

sql_df.head()
```

```python
print(f"Rows: {sql_df.shape[0]}")
print(f"Columns: {sql_df.shape[1]}")
```

```python
sql_df.info()
```

```python
conn = sqlite3.connect("../database/churn.db")

print("SQLite Database Connected Successfully!")
```

```python
query = """
SELECT *
FROM customer_churn
LIMIT 5;
"""

pd.read_sql(query, conn)
```

```python
sql_df.to_sql(
    name="customer_churn_data",
    con=conn,
    if_exists="replace",
    index=False
)

print("Table Created Successfully!")
```

```python
query = """
SELECT *
FROM customer_churn_data
LIMIT 5;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT COUNT(*) AS Total_Customers
FROM customer_churn_data;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT name
FROM sqlite_master
WHERE type='table';
"""

pd.read_sql(query, conn)
```

```python
conn.close()
```
